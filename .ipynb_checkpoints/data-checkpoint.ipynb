```python
adDif = DataFrame[DataFrame['price'] <= 500]
```

```python
sns.distplot(adDif['price']) ##sns.histplot(adDif['price'])
```

```python
adDif['price'].mean()
```

```python
### whate is the distribution of the flat price based on neighbourhood group ?
```

```python
plt.figure(figsize=(9,7))
sns.violinplot(data=adDif, x="neighbourhood_group", y="price")
```

```python
### plot on map
```

```python
plt.figure(figsize=(11, 7))
plt.scatter(adDif['longitude'], adDif['latitude'], c=adDif['price'], cmap='inferno', alpha=0.8)
plt.colorbar()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot with Price Color Map')
plt.show()
```
